```python
def clean_it(line):
    return [x for x in line if len(x)>=1]
clean_it(lol)
```

```python
tokens = [clean_it(re.split('[^a-z]', line)) for line in sents]
all_words = [item for sublist in tokens for item in sublist]
all_words
```

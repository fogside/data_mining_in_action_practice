```python
from sklearn.datasets import load_boston
data = load_boston().data
target = load_boston().target
print "data", data.shape
print "target", target.shape
```

```python
from sklearn.preprocessing import scale
data = scale(data)
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score

kf = KFold(n = 178 ,n_folds = 5, random_state = 42, shuffle = True)

work_range = np.linspace(1,10,200)
RES = {i:i for i in work_range}

for param in work_range:
    kreg = KNeighborsRegressor(n_neighbors=5, weights="distance", metric = "minkowski", p = param)
    scores = cross_val_score(kreg, data, target, cv=kf, scoring='mean_squared_error')
    RES[param] = np.mean(scores)
    
    
from operator import itemgetter

sorted(RES.items(), key=itemgetter(1))
```
